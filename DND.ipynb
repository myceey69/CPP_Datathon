```python
country = (
    df.groupby("Country", as_index=False)
      .agg(
          avg_gdp=("GDP", "mean"),
          avg_mortality=("Mortality Rate (%)", "mean"),
          avg_health_access=("Healthcare Access (%)", "mean"),
          avg_doctors=("Doctors per 1000", "mean"),
          avg_beds=("Hospital Beds per 1000", "mean"),
          avg_urban=("Urbanization Rate (%)", "mean")
      )
)
```

```python
corr = country["avg_gdp"].corr(country["avg_mortality"])
print("Correlation (GDP vs Mortality):", corr)

plt.figure()
plt.scatter(country["avg_gdp"], country["avg_mortality"])
plt.xlabel("Average GDP")
plt.ylabel("Average Mortality Rate (%)")
plt.title("GDP vs Mortality Rate by Country")
plt.show()
```

```python


x = country["avg_gdp"].values
y = country["avg_mortality"].values


b, a = np.polyfit(x, y, 1)
print("\nRegression: mortality = a + b*GDP")
print("a (intercept):", a)
print("b (slope):", b)


plt.figure()
plt.scatter(x, y)
plt.plot(x, a + b*x)
plt.xlabel("Average GDP")
plt.ylabel("Average Mortality Rate (%)")
plt.title("GDP vs Mortality with Regression Line (Country-level)")
plt.show()
```

```python
country_cost_access = (
    df.groupby("Country", as_index=False)
      .agg(
          avg_treatment_cost=("Average Treatment Cost (USD)", "mean"),
          avg_health_access=("Healthcare Access (%)", "mean")
      )
)

country_cost_access.head()
```

```python
x = country_cost_access["avg_treatment_cost"]
y = country_cost_access["avg_health_access"]

# Fit linear regression line
m, b = np.polyfit(x, y, 1)

plt.figure(figsize=(6,4))
plt.scatter(x, y)

# Plot regression line
plt.plot(x, m*x + b)

plt.xlabel("Average Treatment Cost (USD)")
plt.ylabel("Average Healthcare Access (%)")
plt.title("Treatment Cost vs Healthcare Access")
plt.show()
```

```python
country_cost_access = (
    df.groupby("Country", as_index=False)
      .agg(
          avg_treatment_cost=("Average Treatment Cost (USD)", "mean"),
          avg_health_access=("Healthcare Access (%)", "mean"),
          avg_gdp=("GDP", "mean")
      )
)

country_cost_access[[
    "avg_treatment_cost",
    "avg_health_access",
    "avg_gdp"
]].corr()
```

```python
from sklearn.metrics import r2_score

y_pred = m*x + b
r2 = r2_score(y, y_pred)

plt.figure(figsize=(6,4))
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.text(min(x), max(y), f"R² = {r2:.3f}")
plt.xlabel("Average Treatment Cost (USD)")
plt.ylabel("Average Healthcare Access (%)")
plt.title("Treatment Cost vs Healthcare Access")
plt.show()
```

```python
country_efficiency = (
    df.groupby("Country", as_index=False)
      .agg(
          avg_recovery=("Recovery Rate (%)", "mean"),
          avg_doctors=("Doctors per 1000", "mean")
      )
)

country_efficiency["recovery_per_doctor"] = (
    country_efficiency["avg_recovery"] /
    country_efficiency["avg_doctors"]
)

country_efficiency.sort_values("recovery_per_doctor", ascending=False).head()
```

```python
country_burden = (
    df.groupby("Country", as_index=False)
      .agg(
          total_dalys=("DALYs", "sum"),
          avg_gdp=("GDP", "mean")
      )
)

country_burden["dalys_per_gdp"] = (
    country_burden["total_dalys"] /
    country_burden["avg_gdp"]
)

country_burden.sort_values("dalys_per_gdp", ascending=False).head()
```
